```python
import pandas as pd

workbook=pd.read_excel('test.xlsx', skiprows=7, thousands=",")
workbook1=pd.read_excel('test.xlsx', '부용초 거래명세서', skiprows=7, thousands=",")
workbook2=pd.read_excel('test.xlsx', '양강초 거래명세서', skiprows=7, thousands=",")
workbook3=pd.read_excel('test.xlsx', '용화초 거래명세서', skiprows=7, thousands=",")
workbook = workbook.append(workbook1)
workbook = workbook.append(workbook2)
workbook = workbook.append(workbook3)
workbook.drop(['Unnamed: 2'], axis = 1, inplace = True)
workbook.drop(['Unnamed: 3'], axis = 1, inplace = True)
workbook.drop(['Unnamed: 4'], axis = 1, inplace = True)
workbook.drop(['Unnamed: 5'], axis = 1, inplace = True)
workbook.drop(['Unnamed: 9'], axis = 1, inplace = True)
workbook.drop(['Unnamed: 11'], axis = 1, inplace = True)
```

```python
workbook
```

```python
workbook['날짜'] = pd.to_datetime(workbook['날짜'], errors='coerce', dayfirst=False, yearfirst=False, utc=None, format=None, exact=True, unit=None, infer_datetime_format=False, origin='unix', cache=True)
workbook['날짜'] = (workbook['날짜'].dt.strftime('%Y-%m-%d'))
workbook2 = workbook.dropna(how="any", axis = 0, subset = ["날짜", "품      명    /    규      격", "단위", "수   량", "원 산 지"])
workbook2 = workbook2.loc[:, ["날짜", "품      명    /    규      격", "단위", "수   량", "원 산 지"]]
workbook2 = workbook2.set_index("날짜")
print(workbook2)
```

```python
workbook2.to_excel('실습결과.xlsx')
```
